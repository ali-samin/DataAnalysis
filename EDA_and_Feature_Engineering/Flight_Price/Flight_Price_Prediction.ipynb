##### use label encoding for total stops 

```python
df['Total_Stops'].mode()
```

```python
# as we convert stops label encoding by 

#     non stop = 0
#      1 stop = 1
#      2 stop = 2
#      3 stop = 3
#      4 stop = 4

# we have a problem here we dont know which value have to replace by nan so we have to calculate most freaquent value and replac
# nan by most frequent value (so we have to calculate most frequent value using mode then we got mode = 1 stop)
# therefore we replace  nan value by 1.
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```
